```python
import pandas as pd
```

```python
df = pd.read_csv('../County_President/countypres_2000-2020.csv')
```

```python
# drop unused columns and nulls
df = df.drop(columns=['mode', 'version', 'office'])
df = df.dropna()
df = df.reset_index()
df.shape
```

```python
# drop counties reporting zero total votes
df = df.loc[df['totalvotes'] != 0]
df.shape
```

```python
# get % of votes each party received in each county
temp = []

for index, row in df.iterrows():
    temp.append(round(row['candidatevotes'] / row['totalvotes'], 4))
df['percentage'] = temp
```

```python
#create new dataframe with desired data config
curr_county = 'AUTAUGA'
new_df = []
temp_dict = {}
for index, row in df.iterrows():
    
    if row['county_name'] == curr_county:
        temp_dict['year'] = row['year']
        temp_dict['county'] = row['county_name']
        temp_dict['state'] = row['state_po']
        temp_dict['county_fips'] = int(row['county_fips'])
        temp_dict['total_no'] = row['totalvotes']
        curr_party = row['party'][0].lower()
        
        if curr_party == 'd':
            temp_dict['dem_no'] = row['candidatevotes']
            temp_dict['dem_pct'] = row['percentage']
        elif curr_party == 'r':
            temp_dict['rep_no'] = row['candidatevotes']
            temp_dict['rep_pct'] = row['percentage']
        elif curr_party == 'l':
            temp_dict['lib_no'] = row['candidatevotes']
            temp_dict['lib_pct'] = row['percentage']
        elif curr_party == 'g':
            temp_dict['grn_no'] = row['candidatevotes']
            temp_dict['grn_pct'] = row['percentage']
        else:
            temp_dict['oth_no'] = row['candidatevotes']
            temp_dict['oth_pct'] = row['percentage']
    else:
        copy = temp_dict.copy()
        new_df.append(copy)
        temp_dict.clear()
        
        temp_dict['year'] = row['year']
        temp_dict['county'] = row['county_name']
        temp_dict['state'] = row['state_po']
        temp_dict['county_fips'] = int(row['county_fips'])
        temp_dict['total_no'] = row['totalvotes']
        curr_party = row['party'][0].lower()

        if curr_party == 'd':
            temp_dict['dem_no'] = row['candidatevotes']
            temp_dict['dem_pct'] = row['percentage']
        elif curr_party == 'r':
            temp_dict['rep_no'] = row['candidatevotes']
            temp_dict['rep_pct'] = row['percentage']
        elif curr_party == 'l':
            temp_dict['lib_no'] = row['candidatevotes']
            temp_dict['lib_pct'] = row['percentage']
        elif curr_party == 'g':
            temp_dict['grn_no'] = row['candidatevotes']
            temp_dict['grn_pct'] = row['percentage']
        else:
            temp_dict['oth_no'] = row['candidatevotes']
            temp_dict['oth_pct'] = row['percentage']
    
    curr_county = row['county_name']
    

new_df = pd.DataFrame(new_df)
```

```python
def get_num_wins(year):
    winning_party = '#'
    curr_state = 'AL'
    total = 0
    dem_total = 0
    rep_total = 0
    lib_total = 0
    grn_total = 0
    oth_total = 0
    max_anomoly = 0
    for index, row in new_df.iterrows():
        if row['year'] > year:
            break
        if row['year'] != year:
            continue
        if row['state'] == curr_state:
            total += row['total_no']
            dem_total += row['dem_no']
            rep_total += row['rep_no']
            lib_total += row['lib_no']
            grn_total += row['grn_no']
            oth_total += row['oth_no']
            
        else:
            max = 0
            if dem_total > max:
                winning_party = 'DEM'
                max = dem_total
            if rep_total > max:
                winning_party = 'REP'
                max = rep_total
            if lib_total > max:
                winning_party = 'LIB'
                max = lib_total
            if grn_total > max:
                winning_party = 'GRN'
                max = grn_total
            if oth_total > max:
                winning_party = 'OTH'
                max = oth_total
                
                
            print(f'Winning party for {curr_state}: {winning_party}\n\
            dem pct: {round(dem_total/total,4)}\n\
            rep pct: {round(rep_total/total,4)}\n\
            lib pct: {round(lib_total/total,4)}\n\
            grn pct: {round(grn_total/total,4)}\n\
            oth pct: {round(oth_total/total,4)}')
                
                
            curr_state = row['state']
            total = 0
            dem_total = 0
            rep_total = 0
            lib_total = 0
            grn_total = 0
            oth_total = 0
            total += row['total_no']
            dem_total += row['dem_no']
            rep_total += row['rep_no']
            lib_total += row['lib_no']
            grn_total += row['grn_no']
            oth_total += row['oth_no']
```

```python
tmp_df = new_df.drop(columns=['lib_no', 'lib_pct', 'grn_no', 'grn_pct', 'oth_no', 'oth_pct', 'dem_pct', 'rep_pct'])
state_data = pd.read_csv('../us-state-ansi-fips.csv')
```

```python
def get_fips(state):
    for index, row in state_data.iterrows():
        if row[' stusps'].strip() == state:
            return row[' st']
    return -1
```

```python
final_df = []
temp_dict = {}
curr_state = 'AL'
for index, row in tmp_df.iterrows():
    if row['state'] == curr_state:
        temp_dict['year'] = row['year']
        temp_dict['state'] = row['state']
        temp_dict['state_fips'] = get_fips(row['state'])
        if 'total_no' in temp_dict:
            temp_dict['total_no'] += row['total_no']
            temp_dict['dem_no'] += row['dem_no']
            temp_dict['rep_no'] += row['rep_no']
        else:
            temp_dict['total_no'] = row['total_no']
            temp_dict['dem_no'] = row['dem_no']
            temp_dict['rep_no'] = row['rep_no']
    else:
        copy = temp_dict.copy()
        final_df.append(copy)
        temp_dict.clear()
        
        temp_dict['year'] = row['year']
        temp_dict['state'] = row['state']
        temp_dict['state_fips'] = get_fips(row['state'])
        if 'total_no' in temp_dict:
            temp_dict['total_no'] += row['total_no']
            temp_dict['dem_no'] += row['dem_no']
            temp_dict['rep_no'] += row['rep_no']
        else:
            temp_dict['total_no'] = row['total_no']
            temp_dict['dem_no'] = row['dem_no']
            temp_dict['rep_no'] = row['rep_no']
            
        curr_state = row['state']
        
        
final_df = pd.DataFrame(final_df)
```

```python
duplicate = []
counter = {}
for index, row in final_df.iterrows():
    k = str(row['year'])+row['state']
    if k in counter:
        counter[k]+=1
        duplicate.append(index)
    else:
        counter[k]=1
```

```python
final_df = final_df.set_index("state_fips")
final_df["state_fips"] = final_df.index
final_df.head()
```

```python
from statsmodels.tsa.ar_model import AutoReg
import math
```

```python
forecasts = {}
# final_df.index = pd.DatetimeIndex(final_df.index).to_period('M')
for group, values in final_df.groupby("state").dem_no:
    # form the model and fit
    model = AutoReg(values, lags=0)
    result = model.fit()

    # predict
    prediction = result.forecast(steps=1)
    
    # store
    forecasts[group] = prediction

    
dem_predictions = pd.DataFrame(forecasts)
```

```python
forecasts = {}
# final_df.index = pd.DatetimeIndex(final_df.index).to_period('M')
for group, values in final_df.groupby("state").rep_no:
    # form the model and fit
    model = AutoReg(values, lags=0)
    result = model.fit()

    # predict
    prediction = result.forecast(steps=1)
    
    # store
    forecasts[group] = prediction

    
rep_predictions = pd.DataFrame(forecasts)
```

```python
plot_df = []
temp_dict = {}

for col in dem_predictions.columns:
    a=0
    b=0
    
    temp_dict['year'] = 2024
    temp_dict['state'] = col
    temp_dict['state_fips'] = get_fips(col)
    
    for item in dem_predictions[col]:
        if not math.isnan(item):
            print(f'{col}: {item}', end='\t')
            a = item
            temp_dict['dem_no'] = item
    for item in rep_predictions[col]:
        if not math.isnan(item):
            print(f'{col}: {item}', end='\t')
            b = item
            temp_dict['rep_no'] = item
    if a > b:
        print(f'DEM wins {col}')
    elif b > a:
        print(f'REP wins {col}')
        
    copy = temp_dict.copy()
    plot_df.append(copy)
    temp_dict.clear()
    
```

```python
plot_df = pd.DataFrame(plot_df)
plot_df = plot_df.set_index("state_fips")
plot_df["state_fips"] = plot_df.index
```
